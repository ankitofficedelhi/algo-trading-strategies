```python
import dask.dataframe as dd
import pandas as pd 
import datetime
import glob
```

```python

all_files = glob.glob(r"C:\Users\ashu4\Downloads\bills/*.csv")
print(f"Found {len(all_files)} files")
print(all_files)
```

```python
df_list = [pd.read_csv(file) for file in all_files]
df = pd.concat(df_list, ignore_index=True)

df_filtered = df[df['TradingSymbol'].str.startswith('NIFTY')].copy()
```

```python
df_filtered['extracted_date'] = df_filtered['TradingSymbol'].apply(lambda x: x.split()[1])
df_filtered['extracted_date'] = pd.to_datetime(df_filtered['extracted_date'], format='%d%b%Y')
```

```python
df_filtered['index'] = df_filtered['TradingSymbol'].apply(lambda x: x.split()[0])
```

```python
print(f"records in date range: {len(df_filtered)}")
```

```python
df_filtered['transaction_amount'] = abs(df_filtered['BuyAmount'].astype(float)) + abs(df_filtered['SellAmount'].astype(float))
```

```python
df_filtered.groupby(['LoginID', 'index', 'extracted_date'], as_index=False)['transaction_amount'].sum()
```

```python
result = df_filtered.groupby(['LoginID']).agg({
    'BuyAmount': 'sum',
    'SellAmount': 'sum',
    'transaction_amount': 'sum'
}).reset_index()
```

```python
result.to_csv(r"C:\Users\ashu4\Downloads\14-11-2025_to_20-11-2025(SENSEX).csv", index=False)
```
