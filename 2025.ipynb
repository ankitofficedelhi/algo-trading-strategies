```python
def option_banknifty_preprocess(df:pd.DataFrame)->pd.DataFrame:
    
    df["date_time"] = pd.to_datetime(df.Date + " " + df.Time)
    print("date time column created")

    df["index"] = df["Ticker"].str.slice(0,9)
    print("index col created")

    df["expire"] = df["Ticker"].str.slice(9,16)
    print("expire col created ")

    df["scrip"] = df["Ticker"].str.slice(16)
    print("scrip col created ")

    df["expiry_date"] = pd.to_datetime(df["expire"])
    print("expire date col created ")

    df = df[["date_time","Date","index","scrip","Open","High","Low","Close","Volume","expiry_date"]]
    df["dte"] = (df.expiry_date -  pd.to_datetime(df.Date)).dt.days
    print("dte col created ")

    df = df[["date_time","index","scrip","Open","High","Low","Close","Volume","expiry_date","dte"]]
    df = df.set_index("date_time",drop=True)
    print("index set")
    
    return df
```

```python
def option_nifty_preprocess(df:pd.DataFrame)->pd.DataFrame:
    
    df["date_time"] = pd.to_datetime(df.Date + " " + df.Time)
    print("date time column created")

    df["index"] = df["Ticker"].str.slice(0,5)
    print("index col created")

    df["expire"] = df["Ticker"].str.slice(5,12)
    print("expire col created ")

    df["scrip"] = df["Ticker"].str.slice(12)
    print("scrip col created ")

    df["expiry_date"] = pd.to_datetime(df["expire"])
    print("expire date col created ")

    df = df[["date_time","Date","index","scrip","Open","High","Low","Close","Volume","expiry_date"]]
    df["dte"] = (df.expiry_date -  pd.to_datetime(df.Date)).dt.days
    print("dte col created ")

    df = df[["date_time","index","scrip","Open","High","Low","Close","Volume","expiry_date","dte"]]
    df = df.set_index("date_time",drop=True)
    print("index set")
    
    return df
```
